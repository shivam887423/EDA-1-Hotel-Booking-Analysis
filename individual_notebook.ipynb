# **Data Defining**

1*.hotel: : Hotel(Resort Hotel or City Hotel)

2.is_canceled : Value indicating if the booking was canceled (1) or not (0)

3.lead_time :* Number of days that elapsed between the entering date of the booking into the PMS and the arrival date*

4.arrival_date_year : Year of arrival date

5.arrival_date_month : Month of arrival date

6.arrival_date_week_number : Week number of year for arrival date

7.arrival_date_day_of_month : Day of arrival date

8.stays_in_weekend_nights : Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel

9.stays_in_week_nights : Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel

10.adults : Number of adults

11.children : Number of children

12.babies : Number of babies

13.meal : Type of meal booked. Categories are presented in standard hospitality meal packages:

14.country : Country of origin.`

15.market_segment : Market segment designation. In categories, the term â€œTAâ€ means â€œTravel Agentsâ€ and â€œTOâ€ means â€œTour Operatorsâ€

16.distribution_channel : Booking distribution channel. The term â€œTAâ€ means â€œTravel Agentsâ€ and â€œTOâ€ means â€œTour Operatorsâ€

17.is_repeated_guest : Value indicating if the booking name was from a repeated guest (1) or not (0)

18.previous_cancellations : Number of previous bookings that were cancelled by the customer prior to the current booking

19.previous_bookings_not_canceled : Number of previous bookings not cancelled by the customer prior to the current booking

20.reserved_room_type : Code of room type reserved. Code is presented instead of designation for anonymity reasons.

21.assigned_room_type : Code for the type of room assigned to the booking.

22.booking_changes : Number of changes/amendments made to the booking from the moment the booking was entered on the PMS until the moment of check-in or cancellation

23.deposit_type : Indication on if the customer made a deposit to guarantee the booking.

24.agent : ID of the travel agency that made the booking

25.company : ID of the company/entity that made the booking or responsible for paying the booking.

26.days_in_waiting_list : Number of days the booking was in the waiting list before it was confirmed to the customer

27.customer_type : Type of booking, assuming one of four categories

28.adr : Average Daily Rate as defined by dividing the sum of all lodging transactions by the total number of staying nights

29.required_car_parking_spaces : Number of car parking spaces required by the customer

30.total_of_special_requests :* Number of special requests made by the customer (e.g. twin bed or high floor)*

31.reservation_status : Reservation last status, assuming one of three categories

*Canceled â€“ booking was canceled by the customer
 *Check-Out â€“ customer has checked in but already departed
 *No-Show â€“ customer did not check-in and did inform the hotel of the reason why
reservation_status_date** : Date at which the last status was set. This
variable can be used in conjunction with the ReservationStatus to understand when was the booking canceled or when did the customer checked-out of the hotel

# **DATA** **CLEANING**


```python
#lets indentify the duplicate values in each columns
duplicate_columns=data[data.duplicated()]
data.count()
```

```python
#lets see duplicate value in rows using Boolean Type
#true means duplicate rows and False means non duplicate rows
data.duplicated().value_counts()
```

```python
#Plotting the graph of non value count Boolean Value
plt.figure(figsize=(5,4))
sns.countplot(x=data.duplicated())
plt.title('Duplicate values in Rows',color='black')
plt.ylabel('Duplicate value',color='green')
plt.xlabel('Boolean Type',color='red')
```

```python
#lets remove the duplicated value and see how much data is left 
data=data.drop_duplicates()
data.shape
```

```python
#There are lots of null values in data.Lets check them
data.isna().sum().sort_values(ascending=False)
```

```python
#Lets fill the integer type data with "0" and string type with string "others"
data['agent'].fillna(0,inplace=True)
data['company'].fillna(0,inplace=True)
data['country'].fillna('others',inplace=True)
data['children'].fillna(0,inplace=True)
data
```

```python
# now lets see the data after removing null value
data.isna().sum().sort_values(ascending=False)
```

```python
# lets add some new columns

data['total_people'] = data['adults'] + data['babies'] +data['children'] # we will use this in later
data.count()
```

# 3.Which Country Has the maximum numbers of bookings?

```python
#1.Which Country visited the most ?
#Lets visualize and try to find the answer
data['country'].value_counts().reset_index().head(10)
data
```

```python
#countries with most booking, month wise

#grouping by countries and making dummy data of month
cntry = pd.get_dummies(data, columns=['arrival_date_month']).groupby('country').sum()

#selecting countries which have more than 27000 arrival_data_day_month
sc = cntry.loc[cntry['arrival_date_day_of_month']>=27000]

#removing prefix from every month
rp = sc.rename(columns = lambda x: x.replace('arrival_date_month_', ''))

#selecting particular columns
sp = rp.loc[:, 'April':'September']

plt.figure(figsize = (18,8))
ax = sns.lineplot(data= sp, dashes=False)
ax.set_xlabel('Countries', fontsize = 20)
ax.set_ylabel('Visit Count', fontsize = 20)
ax.set_title('Month Wise Visiting Countries', fontsize=30)
plt.show()
```

**CONCLUSION**

Portugal has the maximum numbers of bookings.

Brazil has the least numbers of bookings.

Country only Portugal and GBR has more resort hotels bookings as compared to city hotels.

Contries like Netherland,Brazil,Itali,Belgium has the least number of city resort to hotel resort.

The average sum of bookings of countries (BEL,ITA,NLD,BRA) is less than the bookings of GBR alone.
